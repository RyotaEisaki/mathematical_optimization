```python
import pandas as pd, networkx as nx
```

```python
dfnd = pd.read_csv('node0.csv') #　頂点情報ファイル
```

```python
dfed = pd.read_csv('edge0.csv') # 辺情報ファイル
```

```python
g = nx.Graph() # グラフの作成
```

```python
for row in dfnd.itertuples(False):
    dc=row._asdict()
    g.add_node(dc['id'],**dc)
```

```python
for row in dfed.itertuples(False):
    dc=row._asdict()
    g.add_edge(dc['node1'],dc['node2'],**dc)
```

```python
nx.draw(g)
```

```python
g.nodes[0]
```

```python
g.edges[0,1]
```

```python
nx.draw_networkx(g) ### 頂点ラベルの表示
```
