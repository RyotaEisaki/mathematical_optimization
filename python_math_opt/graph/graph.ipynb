```python
%matplotlib inline
import networkx as nx
import matplotlib.pyplot as plt
```

```python
G=nx.Graph()
vlist=[1,2,3,4]
elist=[(1,1),(1,1),(2,1),(1,3),(2,3),(2,4),(3,4)]
G.add_nodes_from(vlist)
G.add_edges_from(elist)
nx.draw_networkx(G,node_color='lightgray',node_size=400)
plt.axis('off')
plt.show()
```

```python
print('virtex list: ',G.nodes())
print('virtex number: ',G.number_of_nodes())
print('neighbors of virtex 1: ', [v for v in nx.all_neighbors(G,1)])
```

```python
DG=nx.DiGraph()
vlist=[1,2,3,4]
elist=[(1,1),(1,1),(2,1),(1,3),(2,3),(2,4),(3,4)]
DG.add_nodes_from(vlist)
DG.add_edges_from(elist)
nx.draw_networkx(DG,node_color='lightgray',node_size=400)
plt.axis('off')
plt.show()
```

```python
print('DG edge list: ', DG.edges())
print('DG edge number: ', DG.number_of_edges())
```

```python
G=nx.MultiGraph()
G.add_edges_from([(1,2),(1,3),(3,1),(2,3),(2,2)])
A=nx.adjacency_matrix(G)
M=nx.incidence_matrix(G)
print('A=',A.toarray())
print('M=',M.toarray())
```

```python
nx.draw_networkx(G,node_color='lightgray',node_size=400)
plt.axis('off')
plt.show()
```
